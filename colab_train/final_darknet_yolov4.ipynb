```python
# yolov4-tiny-obj.cfg配置文件修改設定值

# 切換目錄
%cd /content/drive/MyDrive/yolov4/darknet/data

# 第6行的值，將64改為64                           batch=64 -> 64  GPU(batch=64, subdivisions=16)、CPU(batch=16, subdivisions=8)、快速學習與練習(batch=8, subdivisions=4)
!sed -i '6s/64/64/' yolo/cfg/yolov4-tiny-obj.cfg
# 第7行的值，將1改為32                           subdivisions=1 -> 16  數值越大可減少每次訓練所需的記憶體適應較小的記憶體與提高訓練的穩定性
!sed -i '7s/1/6/' yolo/cfg/yolov4-tiny-obj.cfg
# 第20行的值，將500200改為6000                       max_batches=500200 -> max(classes*2000, 6000) = max(2*2000, 6000) = 6000  數值越大訓練時間越長越能優化，但數據集精簡時可能過擬合
!sed -i '20s/500200/6000/' yolo/cfg/yolov4-tiny-obj.cfg
# 第22行的值，將400000,450000改為4800,5400                 steps=400000,450000 -> max_batches*0.8,max_batches*0.9 = 6000*0.8,6000*0.9 = 4800,5400
!sed -i '22s/400000,450000/480,540/' yolo/cfg/yolov4-tiny-obj.cfg
# 第212行的值，將255改為21                         filters=255 -> filters=(classes+5)*3 = (2+5)*3 = 21
!sed -i '212s/255/33/' yolo/cfg/yolov4-tiny-obj.cfg
# 第220行的值，將80改為2                          classes=80 -> 2 ['apple', 'strawberry']
!sed -i '220s/80/6/' yolo/cfg/yolov4-tiny-obj.cfg
# 第263行的值，將255改為21                         filters=255 -> 21
!sed -i '263s/255/33/' yolo/cfg/yolov4-tiny-obj.cfg
# 第269行的值，將80改為2                          classes=80 -> 2
!sed -i '269s/80/6/' yolo/cfg/yolov4-tiny-obj.cfg

# filters=(classes+5)*3 這個公式的意思是：
# "classes"：表示要檢測的物件類別數。
# "+5"：表示每個bounding box需要預測的額外5個參數，分別是：x、y、w、h和objectness score。
# "*3"：表示每個網格單元預測3個bounding box（YOLOv4中每個網格單元預測多個bounding box）。
```

使用者設定驗證集的百分比 percentage_valid 以及測試集的百分比 percentage_test

```python
# 建立PYTHON檔案，將資料集依訓練與驗證設定比例轉存檔案至設定的照片與驗證資料夾，並寫入 train.txt 和 valid.txt

# 切換目錄
%cd /content/drive/MyDrive/yolov4/darknet

# 定義py檔案路徑
file_path = 'process.py'

# 定義要寫入的程式碼
code = """
# 匯入glob模組，用於文件系統操作
import glob

# 匯入os模組，用於操作系統相關功能
import os

# 匯入shutil模組，用於高階檔案操作
import shutil

# 用於驗證集百分比，20%的驗證集比例通常是一個較好的平衡點
percentage_valid = 20

# 用於測試集百分比，確認執行分配照片後test資料夾裡至少有一張照片
percentage_test = 10

# percentage_train = 100 - percentage_valid - percentage_test

current_dataset = 'data/yolo/cfg/dataset'
current_labels = 'data/yolo/cfg/dataset'
current_train = 'data/yolo/cfg/train'
current_train_file = 'data/yolo/cfg/train.txt'
current_valid = 'data/yolo/cfg/valid'
current_valid_file = 'data/yolo/cfg/valid.txt'
current_test = 'data/yolo/cfg/test'
current_test_file = 'data/yolo/cfg/test.txt'

# 建立訓練照片路徑檔 train.txt
file_train = open(current_train_file, 'w')
# 建立驗證照片路徑檔 valid.txt
file_valid = open(current_valid_file, 'w')
# 建立測試照片路徑檔 test.txt
file_test = open(current_test_file, 'w')

# 取得dataset所有 .jpg 檔案的路徑
jpg_files = glob.glob(os.path.join(current_dataset, '*.jpg'))

# 計算 .jpg 檔案的總數
total_jpg_files = len(jpg_files)

# 依訓練與驗證的照片量百分比分別寫入照片路徑至 train.txt 和 valid.txt
num_valid = round(percentage_valid / 100 * total_jpg_files)
num_test = round(percentage_test / 100 * total_jpg_files)

counter = 1
# 預設照片副檔名須為.jpg
for pathAndFilename in glob.iglob(os.path.join(current_dataset, '*.jpg')):
    title, ext = os.path.splitext(os.path.basename(pathAndFilename))

    if counter <= num_valid:
        file_valid.write(current_valid + '/' + title + '.jpg' + '\\n')
        if not os.path.exists(current_valid + '/' + title + '.jpg'):
            shutil.copyfile(current_dataset + '/' + title + '.jpg', current_valid + '/' + title + '.jpg')
            shutil.copyfile(current_labels + '/' + title + '.txt', current_valid + '/' + title + '.txt')
    elif counter <= num_test + num_valid:
        file_test.write(current_test + '/' + title + '.jpg' + '\\n')
        if not os.path.exists(current_test + '/' + title + '.jpg'):
            shutil.copyfile(current_dataset + '/' + title + '.jpg', current_test + '/' + title + '.jpg')
            shutil.copyfile(current_labels + '/' + title + '.txt', current_test + '/' + title + '.txt')
    else:
        file_train.write(current_train + '/' + title + '.jpg' + '\\n')
        if not os.path.exists(current_train + '/' + title + '.jpg'):
            shutil.copyfile(current_dataset + '/' + title + '.jpg', current_train + '/' + title + '.jpg')
            shutil.copyfile(current_labels + '/' + title + '.txt', current_train + '/' + title + '.txt')
    counter = counter + 1
"""

# 寫入程式碼到檔案process.py
with open(file_path, 'w') as file:
    file.write(code)

# 顯示檔案已成功建立並寫入程式碼的訊息
print(f"檔案已成功建立並寫入程式碼到 {file_path}")
```

```python
# 切換目錄
%cd /content/drive/MyDrive/yolov4/darknet

# 定義py檔案路徑
file_path = 'process_update.py'

# 定義要寫入的程式碼
code = """
# 匯入glob模組，用於文件系統操作
import glob

# 匯入os模組，用於操作系統相關功能
import os

current_train = 'data/yolo/cfg/train'
current_train_file = 'data/yolo/cfg/train.txt'
current_valid = 'data/yolo/cfg/valid'
current_valid_file = 'data/yolo/cfg/valid.txt'
current_test = 'data/yolo/cfg/test'
current_test_file = 'data/yolo/cfg/test.txt'

# 建立訓練照片路徑檔 train.txt
file_train = open(current_train_file, 'w')
# 建立驗證照片路徑檔 valid.txt
file_valid = open(current_valid_file, 'w')
# 建立測試照片路徑檔 valid.txt
file_test = open(current_test_file, 'w')

# 寫入照片路徑至 train.txt ，預設照片副檔名須為.jpg
for pathAndFilename in glob.iglob(os.path.join(current_train, '*.jpg')):
    title, ext = os.path.splitext(os.path.basename(pathAndFilename))
    file_train.write(current_train + '/' + title + '.jpg' + '\\n')

# 寫入照片路徑至 valid.txt，預設照片副檔名須為.jpg
for pathAndFilename in glob.iglob(os.path.join(current_valid, '*.jpg')):
    title, ext = os.path.splitext(os.path.basename(pathAndFilename))
    file_valid.write(current_valid + '/' + title + '.jpg' + '\\n')

# 寫入照片路徑至 test.txt，預設照片副檔名須為.jpg
for pathAndFilename in glob.iglob(os.path.join(current_test, '*.jpg')):
    title, ext = os.path.splitext(os.path.basename(pathAndFilename))
    file_test.write(current_test + '/' + title + '.jpg' + '\\n')
"""

# 寫入程式碼到檔案process_update.py
with open(file_path, 'w') as file:
    file.write(code)

# 顯示檔案已成功建立並寫入程式碼的訊息
print(f"檔案已成功建立並寫入程式碼到 {file_path}")
```

建立 detect_test_images.py 檔案，列舉 data/yolo/cfg/test.txt 檔照片使用訓練完成的權重檔進行物件偵測所有照片 閾值設定為0.25

```python
# 建立detect_test_images.py，列舉 data/yolo/cfg/test.txt 檔照片使用訓練完成的權重檔進行物件偵測照片 閾值設定為0.25

# 切換目錄
%cd /content/drive/MyDrive/yolov4/darknet

# 定義py檔案路徑
file_path = 'detect_test_images.py'

# 定義要寫入的程式碼
code = """
import os

# 設定檔案路徑
test_file_path = 'data/yolo/cfg/test.txt'
darknet_command_template = './darknet detector test data/yolo/cfg/obj.data data/yolo/cfg/yolov4-tiny-obj.cfg data/yolo/cfg/training/yolov4-tiny-obj_final.weights {} -thresh 0.25'

# 讀取 test.txt 檔案
with open(test_file_path, 'r') as file:
    image_paths = file.readlines()

# 去除每行的換行符號
image_paths = [path.strip() for path in image_paths]

# 對每張照片進行偵測
for image_path in image_paths:
    command = darknet_command_template.format(image_path)
    os.system(command)
"""

# 寫入程式碼到檔案 detect_test_images.py
with open(file_path, 'w') as file:
    file.write(code)

# 顯示檔案已成功建立並寫入程式碼的訊息
print(f"檔案已成功建立並寫入程式碼到 {file_path}")
```
